```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("twitter.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("Date")["Close"].sum()
print(grouped_by)
plt.plot(grouped_by.index,grouped_by.values)
plt.xlabel("Date",fontsize=20)
plt.ylabel("Close",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("twitter.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("Date")["Close"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel("Date",fontsize=20)
plt.ylabel("Close",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("twitter.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("High")["Low"].sum()
print(grouped_by)
plt.scatter(grouped_by.index,grouped_by.values)
plt.xlabel("High",fontsize=20)
plt.ylabel("Low",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("twitter.xlsx")
df=pd.DataFrame(data)
print(df.head())
plt.hist(df["Volume"])
plt.xlabel("Volume",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```
